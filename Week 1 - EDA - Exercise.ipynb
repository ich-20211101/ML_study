```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ich-20211101/appliedMultivariateStatisticalAnalysis/refs/heads/main/wine.csv')
df.head()
```

```python
plt.title('Count of Number of Each Wine')
sns.countplot(x='Wine', data=df)
```

```python
corr_matrix = df.corr(method = 'pearson')
corr_matrix
```

```python
corr_matrix[['Wine']].unstack().sort_values(ascending=False)
```

```python
corr_matrix.style.background_gradient(cmap='coolwarm').format(precision = 2)
```

```python
sns.boxplot(x = 'Wine', y = 'Alcohol', data = df)
```

```python
sns.scatterplot(x = 'Alcohol', y = 'Proline', data = df)
```

```python
sns.regplot(x = 'Alcohol', y = 'Proline', data = df)
```

```python
df.hist(bins = 20, figsize = (10, 10))
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```
